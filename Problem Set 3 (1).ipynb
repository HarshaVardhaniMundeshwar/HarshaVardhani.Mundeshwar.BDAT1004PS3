Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
# Step 1. Import the necessary libraries
import pandas as pd

# Step 2. Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')  # Step 3. Assign it to a variable called users

# Step 4. Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean age per occupation:")
print(mean_age_per_occupation)

# Step 5. Discover the Male ratio per occupation and sort it from the most to the least
male_ratio_per_occupation = (
    users.groupby('occupation')['gender']
    .apply(lambda x: (x == 'M').mean())
    .sort_values(ascending=False)
)
print("\nMale ratio per occupation (sorted):")
print(male_ratio_per_occupation)

# Step 6. For each occupation, calculate the minimum and maximum ages
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMinimum and maximum ages per occupation:")
print(min_max_age_per_occupation)

# Step 7. For each combination of occupation and sex, calculate the mean age
mean_age_per_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean age per occupation and gender:")
print(mean_age_per_occupation_gender)

# Step 8. For each occupation, present the percentage of women and men
gender_percentage_per_occupation = users.groupby('occupation')['gender'].value_counts(normalize=True).unstack()
gender_percentage_per_occupation = gender_percentage_per_occupation.fillna(0) * 100
gender_percentage_per_occupation.columns = ['female_percentage', 'male_percentage']
print("\nPercentage of women and men per occupation:")
print(gender_percentage_per_occupation)
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Step 1: Import the dataset
file_path = r'Downloads/wind.txt'  # Update with the correct path
data = pd.read_csv(file_path, sep=r'\s+', na_values='NaN')

# Step 2: Rename columns
data.columns = ['Yr', 'Mo', 'Dy'] + list(data.columns[3:])

# Step 3: Fix the year column (convert two-digit years)
data['Yr'] = data['Yr'].apply(lambda x: 1900 + x if x < 100 else x)

# Step 4: Create a date string and convert to datetime
data['Date_String'] = data['Yr'].astype(str) + '-' + data['Mo'].astype(str) + '-' + data['Dy'].astype(str)
data['Date'] = pd.to_datetime(data['Date_String'], format='%Y-%m-%d', errors='coerce')

# Step 5: Identify and drop invalid dates
data = data.dropna(subset=['Date'])

# Step 6: Set the datetime column as the index
data.set_index('Date', inplace=True)
data.drop(columns=['Yr', 'Mo', 'Dy', 'Date_String'], inplace=True)

# Step 7: Compute missing values for each location
missing_values = data.isna().sum()
print("\nMissing values per location:")
print(missing_values)

# Step 8: Compute total non-missing values
total_non_missing = data.notna().sum().sum()
print("\nTotal non-missing values:", total_non_missing)

# Step 9: Calculate the mean windspeed over all locations and times
overall_mean = data.mean().mean()
print("\nOverall mean windspeed:", overall_mean)

# Step 10: Create loc_stats DataFrame for location statistics
loc_stats = data.describe().T[['min', 'max', 'mean', 'std']]
print("\nLocation statistics:")
print(loc_stats)

# Step 11: Create day_stats DataFrame for daily statistics
day_stats = data.agg(['min', 'max', 'mean', 'std'], axis=1).rename(columns={
    'min': 'Daily Min', 'max': 'Daily Max', 'mean': 'Daily Mean', 'std': 'Daily Std'
})
print("\nDaily statistics:")
print(day_stats.head())

# Step 12: Average windspeed in January for each location
january_data = data[data.index.month == 1]
january_mean = january_data.mean()
print("\nJanuary mean windspeed for each location:")
print(january_mean)

# Step 13: Downsample to yearly frequency (year-end)
yearly_data = data.resample('YE').mean()
print("\nYearly data:")
print(yearly_data.head())

# Step 14: Downsample to monthly frequency (month-end)
monthly_data = data.resample('ME').mean()
print("\nMonthly data:")
print(monthly_data.head())

# Step 15: Downsample to weekly frequency
weekly_data = data.resample('W').mean()
print("\nWeekly data:")
print(weekly_data.head())

# Step 16: Weekly statistics for the first 52 weeks (from Jan 2, 1961)
weekly_stats = weekly_data.iloc[:52].agg(['min', 'max', 'mean', 'std'], axis=1).rename(columns={
    'min': 'Weekly Min', 'max': 'Weekly Max', 'mean': 'Weekly Mean', 'std': 'Weekly Std'
})
print("\nWeekly statistics for the first 52 weeks:")
print(weekly_stats)
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Step 1: Import the dataset from the given URL
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')

# Step 2: Assign the dataset to a variable called chipo (already done in step 1)

# Step 3: See the first 10 entries
print("First 10 entries:")
print(chipo.head(10))

# Step 4: What is the number of observations in the dataset?
num_observations = chipo.shape[0]
print(f"\nNumber of observations: {num_observations}")

# Step 5: What is the number of columns in the dataset?
num_columns = chipo.shape[1]
print(f"Number of columns: {num_columns}")

# Step 6: Print the name of all the columns
print("\nColumn names:", chipo.columns.tolist())

# Step 7: How is the dataset indexed?
print("\nIndex of the dataset:", chipo.index)

# Step 8: Which was the most-ordered item?
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print(f"\nMost ordered item: {most_ordered_item}")

# Step 9: For the most-ordered item, how many items were ordered?
most_ordered_item_quantity = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print(f"Quantity of most ordered item ({most_ordered_item}): {most_ordered_item_quantity}")

# Step 10: What was the most ordered item in the choice_description column?
most_ordered_choice_description = chipo['choice_description'].value_counts().idxmax()
print(f"\nMost ordered choice description: {most_ordered_choice_description}")

# Step 11: How many items were ordered in total?
total_items_ordered = chipo['quantity'].sum()
print(f"Total items ordered: {total_items_ordered}")

# Step 12: 
# • Turn the item price into a float
chipo['item_price'] = chipo['item_price'].replace([r'[\$,]'], '', regex=True).astype(float)

# • Check the item price type
print(f"\nItem price type after conversion: {chipo['item_price'].dtype}")

# • Create a lambda function and change the type of item price
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x))

# • Check the item price type again
print(f"Item price type after applying lambda function: {chipo['item_price'].dtype}")

# Step 13: How much was the revenue for the period in the dataset?
revenue = (chipo['item_price'] * chipo['quantity']).sum()
print(f"\nTotal revenue: ${revenue:.2f}")

# Step 14: How many orders were made in the period?
num_orders = chipo['order_id'].nunique()
print(f"Number of unique orders: {num_orders}")

# Step 15: What is the average revenue amount per order?
average_revenue_per_order = revenue / num_orders
print(f"Average revenue per order: ${average_revenue_per_order:.2f}")

# Step 16: How many different items are sold?
num_different_items = chipo['item_name'].nunique()
print(f"\nNumber of different items sold: {num_different_items}")
```

Question 9


Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset from the given file path
file_path = 'Downloads/roman-emperor-reigns.csv'  # Your file path
data = pd.read_csv(file_path)

# Step 3: Clean column names (remove any leading/trailing spaces)
data.columns = data.columns.str.strip()

# Step 4: Check the 'Cause_of_Death' column and create a new column for 'Assassinated'
# We'll assume that the term 'assassinated' is in the 'Cause_of_Death' column if they were assassinated
data['Assassinated'] = data['Cause_of_Death'].str.contains('assassinated', case=False, na=False)

# Step 5: Calculate the number of assassinated and non-assassinated emperors
assassinated_count = data['Assassinated'].sum()
non_assassinated_count = len(data) - assassinated_count

# Step 6: Create a pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [assassinated_count, non_assassinated_count]
colors = ['red', 'green']

# Step 7: Plot the pie chart
plt.figure(figsize=(7, 7))  # Ensure the pie chart is circular
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'edgecolor': 'black'})

# Step 8: Title the chart
plt.title('Fraction of Roman Emperors Assassinated')

# Step 9: Display the chart
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```
